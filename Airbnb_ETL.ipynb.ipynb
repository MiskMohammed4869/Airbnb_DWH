# Airbnb Data Warehouse - ETL Pipeline

## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import glob
import os
```

## 2. Load Raw Datasets

```python
files = glob.glob("data/raw/*.csv")

dfs = []

for file in files:
    df = pd.read_csv(file)

    filename = os.path.basename(file).replace(".csv","")

    city, day = filename.split("_",1)

    df["city"] = city.title()
    df["day_type"] = "weekend" if "weekend" in day else "weekday"

    dfs.append(df)

airbnb = pd.concat(dfs, ignore_index=True)
```

## 6. Remove Unnecessary Columns

```python
airbnb.drop(columns=["Unnamed: 0"], inplace=True)
```

## 7. Data Type Validation

```python
airbnb.dtypes
```

## 8. Convert Data Types

```python
airbnb["person_capacity"] = airbnb["person_capacity"].astype(int)
airbnb["cleanliness_rating"] = airbnb["cleanliness_rating"].astype(int)

airbnb["multi"] = airbnb["multi"].astype(bool)
airbnb["biz"] = airbnb["biz"].astype(bool)
airbnb["host_is_superhost"] = airbnb["host_is_superhost"].astype(bool)
```

```python
airbnb.info()
```

## 9. Save Clean Dataset

```python
airbnb.to_csv("data/processed/airbnb.csv", index=False)
```
